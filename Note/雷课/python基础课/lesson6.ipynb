```python
#打开文件
fp = open('sample.txt', 'r')#打开一个文件，r读文件
content = fp.read()#把文件内容全部读出来
print(content)
fp.close()
```

```python
with open('sample.txt', 'r') as fp:#下面读语句块执行完，自动关闭文件
    content = fp.read()
    print(content)
```

```python
with open('sample.txt', 'r') as fp:#下面读语句块执行完，自动关闭文件
    content = fp.readline()#读文件读一行内容
    print(content)
```

```python
with open('sample.txt', 'r') as fp:#下面读语句块执行完，自动关闭文件
    content = fp.readlines()#读文件的所有行，按行读，构建list
    print(content)
```

```python
#写文件
with open('test.txt', 'w') as fp:#打开文件，写文件 w表示已经存在的文件清空再写，不存在的文件自动创建
    fp.write('this is a new line')
```

```python
#写文件 追加内容
with open('test.txt', 'a') as fp:#a表示追加内容
    fp.write('\nthis is a new line')
```

```python
# 图片等二进制文件 上面都是文本文件
with open('a.jpeg', 'rb') as fp:#rb表示打开二进制文件读
    content = fp.read()
    print(content)
```

### 例子

```python
#按行打印一个list
def printlist(list1):
    for line in list1:
        print(line[0:-1])

#按照姓名返回年龄
def get_age_by_name(name, list1):
    for line in list1:
        strlist = line.split(',')
        if (strlist[3].find(name) != -1):
            return strlist[6]
    return None


with open('titanic.csv', 'r') as fp:#下面读语句块执行完，自动关闭文件
    contentlist = fp.readlines()#读文件的所有行，按行读，构建list
    print(get_age_by_name('Nasser', contentlist))

with open('female.csv', 'w') as fp:
    fp.write(contentlist[0])#表头
    for line in contentlist:
        if line.find('female') != -1:
            fp.write(line)
```

```python
dic = {}
with open('titanic.csv', 'r') as fp:#下面读语句块执行完，自动关闭文件
    contentlist = fp.readlines()#读文件的所有行，按行读，构建list
    contentlist.remove(contentlist[0])
    for item in contentlist:
        strlist = item.split(',')
        dic.setdefault(strlist[0], strlist[1])
print(dic)
```
