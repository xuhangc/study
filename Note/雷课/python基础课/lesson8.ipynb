### 小结

绘图的流程

1. 明确问题

2. 选择图形

3. 准备数据

4. 绘制图形

```python
x = [1,2,3,4,5,6,7]
y1 = [27,25,26,29,24,26,28]
y2 = [13,12,17,10,14,16,18]

plt.bar(x, y1, label='high')
plt.bar(x, y2, label='low')

plt.title('bj temp')
plt.xlabel('day')
plt.ylabel('temp')

plt.legend()
plt.show()
```

小贴士

饼图：将份额最大的放在12点方向，顺时针放置第二大份额的部分，以此类推

线状图：用实线

面积图：使用透明效果来确保用户可以看到全部数据

散点图：增加趋势线

柱状图：柱子间的间隔最好调整为柱子宽度的1/2

选择合适的图表，让数据对比更明显直接。柱状图比饼状图在视觉上更易于比较。

确保任何呈现都是准确的。

清楚标明各个图形表示的数据，避免用与主要数据不相关的颜色、形状干扰视觉。

将同类数据归类，简化色彩，帮助用户更快理解数据

要客观反映真实数据，不要随便截断坐标轴，否则视觉感受和实际数据相差很大
