图形属性设置

2.1 标题、标签和图例

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]
plt.plot(x,y)
plt.title('Title')
plt.show()
```

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]
plt.plot(x,y)

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.show()
```

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]
plt.plot(x,y,label='bj')

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.legend()

plt.show()
```

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]

#显示中文字符
plt.rcParams['font.sans-serif'] = ['SimHei'] # 步骤一（替换sans-serif字体）
plt.rcParams['axes.unicode_minus'] = False # 步骤二（解决坐标轴负数的负号显示问题）

plt.plot(x,y,label='北京')

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.legend()

plt.show()
```

2.3 设置图像大小

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]

plt.figure(figsize=(6,4), dpi=100)

plt.plot(x,y,label='bj')

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.legend()

plt.show()
```

2.4 设置颜色和线形状
2.5 添加文字说明

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]

plt.figure(figsize=(6,4), dpi=100)

plt.plot(x,y,label='bj',color='g',linestyle='--',marker='o')

plt.text(2,5.2,s="min value",fontsize=10)

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.legend()

plt.show()
```

2.6 保存图像

```python
x = [0,2,4,6,8]
y = [1,5,3,9,7]

plt.figure(figsize=(6,4), dpi=100)

plt.plot(x,y,label='bj',color='g',linestyle='--',marker='o')

plt.text(2,5.2,s="min value",fontsize=10)

plt.title('Title')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.legend()

plt.savefig('five.jpg')

plt.show()
```
